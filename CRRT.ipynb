```python
import numpy as np

sz=df['gender'].size

# shuffle data
rnd=np.random.rand(1,sz)
df['rnd']=list(rnd[0])
df=df.sort_values('rnd')
# split to training 80% and testing 20%
tr=df[:int(sz*0.80)] #training
ts=df[int(sz*0.80):] # testing

#print(sz)
ts
```

```python
from sklearn.metrics import auc
from sklearn.metrics import roc_curve
from sklearn import metrics

from scipy.stats import pearsonr

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.svm import LinearSVC

from matplotlib import pyplot as plt
```

```python
res_l=[]
for i in range (10,151, 10):
    #n_estimators is number of trees
    rfx=RandomForestClassifier(n_estimators=i)
    rfx.fit(tr[cols],tr['crrt'])
    probs_rfx=rfx.predict_proba(ts[cols])
    fpr_rfx, tpr_rfx, thresholds_rfx = roc_curve(ts['crrt'],probs_rfx[:,1])
    auc_rfx=auc(fpr_rfx,tpr_rfx)
    print('with', i, 'trees', ', the auc is :', auc_rfx)
    res_l.append(auc_rfx)
plt.plot(range(10,151, 10),res_l)
plt.show()
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
rfx=RandomForestClassifier(n_estimators=100)
rfx.fit(tr[cols],tr['crrt'])
probs_rfx=rfx.predict_proba(ts[cols])
fpr_rfx, tpr_rfx, thresholds_rfx = roc_curve(ts['crrt'],probs_rfx[:,1])
auc_rfx=auc(fpr_rfx, tpr_rfx)
plt.plot(fpr_rfx, tpr_rfx, label='RandomForestClassifier')
print('RandomForestClassifier with', i, 'trees', ', the auc is :', auc_rfx)

expected = df['crrt']
predicted = rfx.predict(df[cols])
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
```

```python
LR = LogisticRegression()
LR.fit(tr[cols],tr['crrt'])
#LR
probs_LR=LR.predict_proba(ts[cols])
fpr_LR, tpr_LR, thresholds_LR = roc_curve(ts['crrt'],probs_LR[:,1])
auc_LR=auc(fpr_LR,tpr_LR)
plt.plot(fpr_LR, tpr_LR, label='LogisticRegression')
print('LR auc: ', auc_LR)


expected = df['crrt']
predicted = LR.predict(df[cols])
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
```

```python
GNB= GaussianNB()
GNB.fit(tr[cols],tr['crrt'])
probs_GNB=GNB.predict_proba(ts[cols])
fpr_GNB, tpr_GNB, thresholds_GNB = roc_curve(ts['crrt'],probs_GNB[:,1])
auc_GNB=auc(fpr_GNB,tpr_GNB)
plt.plot(fpr_GNB, tpr_GNB, label='GaussianNB')
#print('fpr_GNB',fpr_GNB)
#print('tpr_GNB',tpr_GNB)
print('GaussianNB auc:', auc_GNB)

expected = df['crrt']
predicted = GNB.predict(df[cols])
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
```

```python
svm = LinearSVC()
probs_svm = svm.fit(tr[cols],tr['crrt']).decision_function(tr[cols])
fpr2svm, tpr2svm, thresholds2svm = roc_curve(tr['crrt'],probs_svm)
auc_SVM=auc(fpr2svm, tpr2svm)
plt.plot(fpr2svm, tpr2svm, label='LinearSVC')
print('SVM auc', auc_SVM)

expected = df['crrt']
predicted = svm.predict(df[cols])
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
```

```python
#import KMeans library to be able to do the KMeans clustering
from sklearn.cluster import KMeans

km=KMeans(n_clusters=5)
km.fit(df[cols])
df['km5']=list(km.labels_)

plt.figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')
plt.title("CRRT vs SOFA clustering")
plt.xlabel("CRRT")
plt.ylabel("SOFA")
plt.scatter(df['crrt'],df['sofa'])
plt.show()
```

```python
plt.figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')
plt.title("KMeans clustering")
plt.xlabel("CRRT")
plt.ylabel("SOFA")
for cluster in range(0,5):
    x=df[df['km5']==cluster]['crrt']
    y=df[df['km5']==cluster]['sofa']
    xx=[]
    for i in x:
        xx.append(i+float(np.random.rand(1,1)/1.5)) #0.5 x movement
    yy=[]
    for i in y:
        yy.append(i+float(np.random.rand(1,1)*5)) #5 y movement
    clr = 'red'
    if cluster ==1:
        clr = 'purple'
    elif cluster == 2:
        clr = 'green'
    elif cluster == 3:
        clr = 'blue'
    elif cluster == 4:
        clr = 'yellow'
    #print(cluster, clr)
    
    plt.scatter(xx,yy, color=clr)
    
plt.show()
```

```python
plt.figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')
plt.title("KMeans clustering")
plt.xlabel("CRRT")
plt.ylabel("AGE")
for cluster in range(0,5):
    x=df[df['km5']==cluster]['crrt']
    y=df[df['km5']==cluster]['age']
    xx=[]
    for i in x:
        xx.append(i+float(np.random.rand(1,1)/1.5)) #0.5 x movement
    yy=[]
    for i in y:
        yy.append(i+float(np.random.rand(1,1)*5)) #5 y movement
    clr = 'red'
    if cluster ==1:
        clr = 'purple'
    elif cluster == 2:
        clr = 'green'
    elif cluster == 3:
        clr = 'blue'
    elif cluster == 4:
        clr = 'yellow'
    #print(cluster, clr)
    
    plt.scatter(xx,yy, color=clr)
    
plt.show()
```

```python
from sklearn.cluster import DBSCAN

cols1 = []
#cols1.append('age')
cols1.append('sofa')
cols1.append('fluid_electrolyte_disorders')
cols1.append('hypotension')
cols1.append('sepsis')
#cols1.append('gcs')

dbs=DBSCAN()
dbs.fit(df[cols1])
df['dbs']=list(dbs.labels_)

#dbs.labels_
#df[cols]

labels = dbs.labels_
#print('uniq values', set(labels))
# Number of clusters in labels, ignoring noise if present.
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print('clusters', n_clusters)

plt.figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')
plt.title("DBSCAN clustering")
plt.xlabel("CRRT")
plt.ylabel("Age")
for cluster in range(-1,4):
    #print('cluster', cluster, df[df['dbs']==cluster]['crrt'].size)
    x=df[df['dbs']==cluster]['crrt']
    y=df[df['dbs']==cluster]['age']
    xx=[]
    for i in x:
        xx.append(i+float(np.random.rand(1,1)/1.5)) #0.5 x movement
    yy=[]
    for i in y:
        yy.append(i+float(np.random.rand(1,1)/1.5)) #0.5 y movement
    clr = 'white' #-1, noise
    if cluster ==0:
        clr = 'purple'
    elif cluster == 1:
        clr = 'green'
    elif cluster == 2:
        clr = 'blue'
    elif cluster == 3:
        clr = 'yellow'
 
    #print(cluster, clr)   
    plt.scatter(xx,yy, color=clr)
    
plt.show()
```
